# 1. Sistemas de Comunicação Digital

Um sistema digital de transmissão de dados contém diversos componentes e etapas que codificam os dados de entrada e os transmitem por um meio físico. A entrada de um sistema digital toma a forma de uma sequência de digitos, que é resultante da codificação de fonte, ou seja, o sinal contínuo  tempo é codificado em uma sequência de símbolos que contém M-bits. <a id="heading--1-1"></a>
## 1.1 Codificador de Linha

A saída digital do codificador de fonte é convertida em pulsos elétricos para fins de transmissão no canal. Há muitas formas de alocar pulsos a dados digitais.
 - *On-off*: O dígito **1** é transmitido por um pulso *p(t)* enquanto que o dígio **0** é transmitido pela ausência de pulso;
 - *polar*: os dígitos **1** e **0** são transmitidos por pulsos *p(t)* e *-p(t)*, respectivamente;
 - Bipolar ou AMI(*alternate makr inversion*): O dígito **0** é transmitido pela ausência de pulso, enquanto que o dígit **1** pode ser transmitido a partir de pulos *p(t)* e *-p(t)*. Nessa codificação, dígitos **1** consecutivos são transmitidos por pulsos de sinal alternado.

 Esses esquemas podem possuir pulsos de meia largura ou de largura completa, onde a amplitude do pulso é mantida constante para pulsos consecutios, sem retornar a 0 enre um pulo e outro. Esses esquemas são denominados **RZ**(*return-to-zero*) e NRZ(non-return-to-zero)

<a id="heading--1-2"></a>
## 1.2 Multiplexador

Em tremos gerais, a capacidade de transmissão de um canal é muito maior do que a largura de banda do sinala  ser transmitido. Para utilizar essa capcidade efetivamente, várias fontes de sinal são combinadas por meio de um multiplexador digital, e essa multiplexação pode ser feita por divisão no tempo, em frequência ou código. Um canal é, portanto, compartilhados por várias mensagens simultâneas.

<a id="heading--1-3"></a>
## 1.3 Repetidor Regenerador

Repetidores regeneradores são utilizados em intervalos regularmente espaçados ao longo de uma linha de transmissão digital para detectar o sinal digital que chega e regenerar novos pulsos “limpos” para transmissão em mais um trecho da linha. Esse processo combate o acúmulo de ruído e distorções de sinal na rota de transmissão. Quando puslso são transmitidos a uma taxa de $R_b$ símbolos por segundo, é necessário obter a informação do sinal de *clock* para amostrar os pulsos que chegam ao receptor. Por exemplo, a retificação do sinal polar RZ produz um sinal de *clock* periódico na frequência desejada. Aplicando esse sinal a um circuito ressonante, obtemos uma senoide que contém a informação de temporização desejada. Utiliza-se um sinal *on-off* a gerado a partir dessa retificação e so sinal transmitido para obter a temporização necessária.

Essa temporização é sensível a o padrão de bits.  Quando o bit **0** é transmitido por meio de ausência de pulso, não haverá sinal na entrada do cricuito ressonante, o que faz a senoide decair, o que pode originar erros na inormação de temporização. Diz-se que o código de linha é transparente quadno o padrão de bits não afeta aprecisão da informação de temporização.


<a id="heading--2"></a>

Discutimos acima como os sinais podem ser transmitidos por diferentes codificações de linha. Cada uma possui vantagens e desvantagens. Em geral, é desejável que o código de linha a ser utilizado tenha as seguintes características

- *Largura de banda* tão pequena quanto possível;
- *Eficiência de potência*;
- *Capacidade de detecção e correção de erros*;
- *Densidade espectral de potência favorável*. Deseja-se que a DSP seja nula ena componente dc do sinal, pois acoplamento e tranformadores ac são utilizados nos repetidores. Alémc disso,potência significativa em frequências baixas é indesejável, pois caus ondulação dc na sequência de pulsos quando acoplamento ac é utilizado;
- *Conteúdo de temporização adequado*;
- *Transprência*.

## 2.1 PSD de códigos de linha

Na discussão a seguir, consideraremos um pulso genérico *p(t)*, cuja tranformada de Fourier é *P(f)*, e denotaremos o símbolo do código de linha no isntante k por $a_k$. Um código de linha gera um trem de pulsos  cstruídos a parir do pulso básico *p(t)*, com amplitude $a_k$, iniciado no tempo $t = kT$, ou seja, o k-ésimo símbolo é transmitodo como $a_kp(t-kT_b)$, onte *T_b* é o período necessário para transmitir um pulso.
Os valores de $a_k$ são aleatórios e definidos pela entrada do codificador $x(t)$ e pelo código utilizado. A saída do codificador de linha $y(t)$ é um sinal modulado em amplitude de pulso(PAM).

Podemos definir a densidade espectral de potência do sinal de saída a partir da equação:

$$
S_y(f) = |p(f)^2| \cdot S_x(f) \tag{2.1}
$$

Para encontrarmos a PSD de $x(t)$, devemos encontrar a sua função de correlação temporal $R_x(\tau)$. Por definição, temos que:

$$
R_{\hat{x}}(\tau) = \lim_{T \to \infty} \frac{1}{T} \int_{-T/2}^{T/2} \hat{x}(t)\hat{x}(t-\tau) dt \tag{2.2}
$$

Consideranto que x(t) possui largura de pulso $\epsilon$ e altura $h_k = \frac{a_k}{\epsilon} \to \infty$. Considerando $\tau < \epsilon$, e sabendo que $R_{\hat{x}}(\tau)$ é uma função par de $\tau$, ou seja, só serão considerados os termos positivos de $\tau$, obtemos:

$$
R_{\hat{x}}(\tau) = \frac{R_0}{\epsilon T}(1/\frac{\tau}{\epsilon})\tag{2.3}
$$

Onde

$$
R_0 = \lim_{N \to \infty} \frac{1}{N} \sum_k a_k^2 \tag{2.4}
$$

Sendo N o número de pulsos transmitidos, e portanto, $R_0$ é a média temporal do quadrado das amplitudes $a_k$ dos pulsos.

Note que para valores de $\tau < \epsilon$, a função de autocorrelação de $x(t)$ é um pulso retangular de amplitude $R_0\epsilon T_b$, largura $2\epsilon$ e centrado em $\tau = 0$. Conforme $\tau$ aumenta, não há sobreposição dos pulsos $x(t)$ e $x(t-\tau)$,e logo a função de autocorrelação é nula. No entando, se continuarmos a aumentar o valor de $\tau$, o sinal atradado se soberpunhará ao próximo pulso transmitido $a_{k+1}$, e assim por diante, de modo que $R_{\hat{x}}(\tau)$ será representado por pulsos triangulares de largura $2\epsilon$, centrados em $nT_b$, com amplitude $R_n/\epsilon T_b$, onde $R_n$ é a média temporal do prduto do pulso $a_k$ e o pulso atrasado $a_{k+n}$

Quando atingimos o limite $\epsilon \to 0$, os pulsos se tornam impulsos. Assim, reescrevemos a função de autocorrelação como:

$$
R_{\hat{x}}(\tau) = \frac{1}{T_b} \sum_{n =-\infty}^{\infty} R_n\delta(\tau -nT_b)\tag{2.4}
$$

A PSD do sinal é dada pela tranformada de Fourier da sua funçaõ de autocorrelação:

$$
S_x(f) = \frac{1}{T_b} \sum_{n = -\infty}^{\infty} R_n e^{-j\omega T_b} \tag{2.6}
$$

Substituindo a expressão da Eq.(2.6) na Eq.(2.1), e lembrando que a função de autocorrelação do sinal é uma função par de $\tau$, obtemos:

$$
S_y(f) = \frac{|P(f)|^2}{T_b}[R_0+2\sum_{n=1}^{\infty} R_n cos(n\omega T_b)]\tag{2.7}
$$

Essa equação nos diz que a PSD de um código de linha é determinada pela escolha do pulso $p(t)$ e pelo correspondente $R_n$
